# Here starts fourth question from project.

```python
'''
Here i have written a function which gives goal differences for all teams and at the same time it gives min and max goal
difference with their teams.
'''
def Goal_diff_count():
    a = []
    for i in range(len(file.Team)):
        diff = file.GoalsFor[i] - file.GoalsAgainst[i]
        a.append(diff)
    for i in range(len(file.Team)):
        if (a[i] == min(a)):
            print('The team with min goal difference is {}'.format(file.Team[i]))
        if (a[i] == max(a)):
            print('The team with max goal difference is {}'.format(file.Team[i]))
    return a
Goal_diff_count()
```

# Here starts sixth question from project.

```python
# To check min and max best positions
minimum = min(file['BestPosition']) 
maximum = max(file['BestPosition'])
minimum, maximum
```

```python
# Here I have grouped the teams with same best positions and I have shown their sum.
for j in range(minimum,maximum+1,1):
    points = 0
    for i in range(len(file)):
        if (file['BestPosition'][i] == j):
            points = points + file['Points'][i]
    print('{}:  {}'.format(j,points))
```
